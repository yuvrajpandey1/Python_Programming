### We have 2 type of storage
1. RAM -> Which is called as Temporary memory
2. ROM -> Which is called as Permanent memory [HDD & SDD]

### Access Modes :
Open(filename,mode)
1. r/r+ -> read only / read & write -> if file does not exit it will give error. Cursor will be at beginning of the file. Contain will remain as it is if file exit.

2. w/w+ -> write / write as well as read -> if file does not exit it will not give error it fill create a file. Cursor will be at beginning of the file.if file exit it will erased all the data inside file.

3. a/a+ -> append / append as well as read -> if file does not exit it will create. Cursor will be at end of the file(append is not but attached).if file exit it will remain as it is.
